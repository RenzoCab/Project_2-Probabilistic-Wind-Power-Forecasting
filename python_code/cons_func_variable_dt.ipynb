```python
theta= 0.01
alpha=10
M=100
N=50
X0=0

start=1e-3
end=5
L=100
dt_test=[0.01,0.1,1,10,100]
r=np.zeros((len(dt_test),L))
s=np.linspace(start,end,L)


for j in range(0,len(dt_test)):
    dt=dt_test[j]
    test_1_param=real(theta,alpha)
    disct_test=disct(N=N,dt=dt,M=M)
    V=gen_path_model_moments(X0,disct_test,test_1_param)
    mm_1=model_moments(disct_test,V )
    
    for i in range(0,L):
        r[j,i]=mm_1.likelihood(np.array([s[i] , alpha]))
```

```python
st=0;
f=200;

plt.plot(s[st:f],r[4,st:f], label='$\\Delta_N$ = '+str(dt_test[4]))
plt.plot(s[st:f],r[3,st:f], label='$\\Delta_N$ = '+str(dt_test[3]))
plt.plot(s[st:f],r[2,st:f], label='$\\Delta_N$ = '+str(dt_test[2]))
plt.plot(s[st:f],r[1,st:f], label='$\\Delta_N$ = '+str(dt_test[1]))
plt.plot(s[st:f],r[0,st:f], label='$\\Delta_N$ = '+str(dt_test[0]))
#plt.ylim([0, 1e5])
#plt.xlim([0, 0.9])
plt.legend()
plt.title('Likelihoods-increasing $\\Delta_N$ for $\\theta = 0.01$, $\\alpha=0.1$ ')
plt.xlabel('$\\theta$')
plt.ylabel('log(Likelihood)')


plt.savefig('dt_increasing_moments_1.pdf')
```
