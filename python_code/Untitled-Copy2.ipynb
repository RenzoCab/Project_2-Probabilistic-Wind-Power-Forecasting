```python
#changin theta

#theta= 0.5
alpha=10
M=1000
N=50
X0=0

start=0.1
end=10
L=100
r=np.zeros((L))
s=np.linspace(start,end,L)


for j in [0.1,0.5,1,2,3]:
    dt=0.1
    theta=j
    test_1_param=real(theta,alpha)
    disct_test=disct(N=N,dt=dt,M=M)
    V=gen_path_model_moments(X0,disct_test,test_1_param)
    mm_1=model_moments(disct_test,V )
    
    for i in range(0,L):
        r[i]=mm_1.likelihood(np.array([s[i] , alpha]))
    plt.plot(s,r, label='theta = '+str(j))
    plt.legend()
    plt.savefig(str(j)+'_theta_close_up_moments.pdf')
```
