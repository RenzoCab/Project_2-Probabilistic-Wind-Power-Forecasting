```python
slope=-0.5
intercept=0.2
axes = plt.gca()
x_vals = np.arange(1,5,1)
y_vals = intercept + slope * x_vals
plt.loglog(10**x_vals, 10**y_vals, '--', label='slope='+str(slope))
plt.loglog(k[:-1,3], np.abs(k[:-1,0]-k[-1,0])/k[-1,0] , '-o' )
plt.xlabel('Number of samples');
plt.ylabel('relative error');
plt.title('Parameter $\\theta_0$');
plt.legend()
plt.savefig('theta_conv_beta.pdf', bbox_inches='tight')
```
