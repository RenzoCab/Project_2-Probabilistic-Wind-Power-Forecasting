```python
from multi_path_base import *
%matplotlib tk 
```

```python
forecast_with_data=np.load('forecast_with_data.npy')
```

```python
interpolation_points=1000
x = np.arange(1,72+1,1)

y = forecast_with_data[80,:,1] #X[j,:]
tck = interpolate.splrep(x, y, s=0)

xnew=np.linspace(1,72,interpolation_points)
p_inter = interpolate.splev(xnew, tck, der=0)
p=p_inter

N=len(p_inter)
dt=1
M_test=1
#real_1 = real(15, 0.1)
#real_1 = real(0.159, 1.40) 
real_1 = real(1e-1,1e-5) 


disct_temp = disct(N,dt,M_test)
```

```python
L=200
beta_X = np.empty((L,len(p_inter)))
beta_X_derivative_tracking = np.empty((L,len(p_inter)))
beta_V = np.empty((L,len(p_inter)))
normal_X = np.empty((L,len(p_inter)))
normal_X_Euler = np.empty((L,len(p_inter)))
normal_X_derivative_tracking = np.empty((L,len(p_inter)))
normal_X_derivative_tracking_Euler = np.empty((L,len(p_inter)))
normal_V = np.empty((L,len(p_inter)))
normal_V_Euler = np.empty((L,len(p_inter)))


for j in range(0,L):
    beta_X[j,:]=gen_X_beta(X0=p_inter[0], disct=disct_temp,real=real_1, forecast=p_inter );
    beta_X_derivative_tracking[j,:]=gen_X_beta_derivative_tracking(X0=p_inter[0], disct=disct_temp,real=real_1, forecast=p_inter );
    beta_V[j,:]=gen_path_beta_robust(X0=0, disct=disct_temp,real=real_1, forecast=p_inter );
    normal_X[j,:]=gen_X_normal(X0=p_inter[0], disct=disct_temp,real=real_1, forecast=p_inter );
    normal_X_derivative_tracking[j,:]=gen_X_normal_derivative_tracking(X0=p_inter[0], disct=disct_temp,real=real_1, forecast=p_inter );
    normal_V[j,:]=gen_path_normal_robust(X0=0, disct=disct_temp,real=real_1, forecast=p_inter );
    normal_X_Euler[j,:]=gen_X_normal_euler(X0=p_inter[0], disct=disct_temp,real=real_1, forecast=p_inter );
    normal_X_derivative_tracking_Euler[j,:]= gen_X_normal_euler_derivative_tracking(X0=p_inter[0], disct=disct_temp,real=real_1, forecast=p_inter );
    normal_V_Euler[j,:]= gen_path_normal_Euler(X0=0, disct=disct_temp,real=real_1, forecast=p_inter );
```

```python
fig = plt.figure(102)
fig.clf()
plt.tight_layout()

#L=200

ax = fig.add_subplot(4,3,7)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,beta_X[j,:], 'k-')
plt.plot(xnew,beta_X[j,:], 'k-', label='simulated forecast')
plt.plot(xnew,p, 'r-', label='forecast')
plt.title('Beta transitions on X without derivative tracking using moments')
plt.legend()

ax = fig.add_subplot(4,3,8)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,beta_X_derivative_tracking[j,:], 'k-')
plt.plot(xnew,p, 'r-')
plt.title('Beta transitions on X with derivative tracking using moments')

ax = fig.add_subplot(4,3,11)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,beta_V[j,:]+p, 'k-')
plt.plot(xnew,p, 'r-')
plt.title('Beta transitions on V using moments')


ax = fig.add_subplot(4,3,4)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,normal_X[j,:], 'k-')
plt.plot(xnew,p, 'r-')
plt.title('Gaussian transitions on X without derivative tracking using moments')

ax = fig.add_subplot(4,3,5)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,normal_X_derivative_tracking[j,:], 'k-')
plt.plot(xnew,p, 'r-')
plt.title('Gaussian transitions on X with derivative tracking using moments')


ax = fig.add_subplot(4,3,10)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,normal_V[j,:]+p, 'k-')
plt.plot(xnew,normal_V[j,:]+p, 'k-', label='simulated forecast')
plt.plot(xnew,p, 'r-', label='forecast')
plt.title('Gaussian transitions on V using moments')

ax = fig.add_subplot(4,3,6)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,normal_X_Euler[j,:], 'k-')
plt.plot(xnew,normal_X_Euler[j,:], 'k-', label='simulated forecast')
plt.plot(xnew,p, 'r-', label='forecast')
plt.title('Transitions on X without derivative tracking using Euler')

ax = fig.add_subplot(4,3,9)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,normal_X_derivative_tracking_Euler[j,:], 'k-')
plt.plot(xnew,normal_X_derivative_tracking_Euler[j,:], 'k-', label='simulated forecast')
plt.plot(xnew,p, 'r-', label='forecast')
plt.title('Transitions on X with derivative tracking using Euler')

ax = fig.add_subplot(4,3,12)
plt.xlim(0, 72)
plt.ylim(0, 1)
for j in range(0,L):
    plt.plot(xnew,normal_V_Euler[j,:]+p, 'k-')
plt.plot(xnew,normal_V_Euler[j,:]+p, 'k-', label='simulated forecast')
plt.plot(xnew,p, 'r-', label='forecast')
plt.title('Transitions on V using Euler')

fig.suptitle('Comparison of Models (V=X-p) \n $\\theta=$' + str(real_1.mu)+ ', $\\alpha=$' +str(real_1.sigma) +', dt=72e-3' ,fontsize=24)

plt.tight_layout()

#plt.savefig('model_comparison.pdf')
#plt.legend()
```
